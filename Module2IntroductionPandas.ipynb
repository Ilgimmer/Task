# Сортировка

```python
# Сортировка по столбцу

df.sort_values(by='movie_name').tail(30)
```

```python
# сортировка по убыванию

df.sort_values(by='grade10', ascending=False).head()
```

```python
# сортировка по нескольким столбцам

df.sort_values(by=['movie_name', 'date', 'grade10'], ascending=[True, True, True]).head(50)
```
